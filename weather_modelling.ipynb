```python
import matplotlib.pyplot as plt
def quadratic_weather_model(day: int, a: float, b: float, c: float) -> float:
    return a * day**2 + b * day + c
def main():
    print("Quadratic Weather Modeling")
    a,b,c=1,-3,2
    num_days=5
    days=list(range(1,num_days+1))
    temperatures=[quadratic_weather_model(day,a,b,c)for day in days]
    max_temp=max(temperatures)
    min_temp=min(temperatures)
    max_day=days[temperatures.index(max_temp)]
    min_day=days[temperatures.index(min_temp)]
    print(f"\nMaximum Temperature: {max_temp:.2f} °C on Day {max_day}")
    print(f"Minimum Temperature: {min_temp:.2f} °C on Day {min_day}")
    day_labels = [f'Day{d}' for d in days]
    plt.figure(figsize=(8, 5))
    plt.plot(days, temperatures, marker='o', linestyle='-', color='darkblue',
    label=f'T = {a}x² + {b}x + {c}')
    plt.scatter(max_day, max_temp, color='red', label=f'Max: {max_temp:.2f}°C (Day {max_day})')
    plt.scatter(min_day, min_temp, color='green', label=f'Min: {min_temp:.2f}°C (Day {min_day})')
    plt.xticks(days, day_labels)
    plt.xlabel('Day')
    plt.ylabel('Temperature (°C)')
    plt.grid(True)
    plt.show()
if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import csv
def quadratic_weather_model(day:int,a:float,b:float,c:float)->float:
  return a*day**2+b*day+c
def main():
  print("Quadratic weather modeling")
  try:
    with open("input.csv",mode='r') as file:
      reader=csv.DictReader(file)
      row=next(reader)
      a=float(row['a'])
      b=float(row['b'])
      c=float(row['c'])
      num_days=int(row['num_days'])
      if num_days<0:
        print("Number of days must be positive")
        return
      days=list(range(1,num_days+1))
      temperatures=[quadratic_weather_model(day,a,b,c) for day in days]
      max_temp=max(temperatures)
      min_temp=min(temperatures)
      max_day=days[temperatures.index(max_temp)]
      min_day=days[temperatures.index(min_temp)]
      print(f"\nMaximum Temperature: {max_temp:.2f} °C on Day {max_day}")
      print(f"Minimum Temperature: {min_temp:.2f} °C on Day {min_day}")
      day_labels = [f'Day{d}' for d in days]
      plt.figure(figsize=(8, 5))
      plt.plot(days, temperatures, marker='o', linestyle='-', color='darkblue',
      label=f'T = {a}x² + {b}x + {c}')
      plt.scatter(max_day, max_temp, color='red', label=f'Max: {max_temp:.2f}°C (Day {max_day})')
      plt.scatter(min_day, min_temp, color='green', label=f'Min: {min_temp:.2f}°C (Day {min_day})')
      plt.xticks(days, day_labels)
      plt.title('Temperature variation')
      plt.xlabel('Day')
      plt.ylabel('Temperature (°C)')
      plt.grid(True)
      plt.show()
  except FileNotFoundError:
    print("Error:'input.csv file not found")
  except ValueError:
    print("Error:csv file contain invalid values")
  except StopIteration:
    print("Error:csv file is empty")
if __name__ == "__main__":
    main()


```

```python
import matplotlib.pyplot as plt
import pandas as pd
def quadratic_weather_model(time,a,b,c):
  temp=a*time**2+b*time+c
  return temp
def main():
  print("Quadratic weather modeling")
  try:
    file_path='sample.csv'
    df=pd.read_csv(file_path)
    time_values=list(range(0,11))
    plt.figure(figsize=(8,6))
    for index,row in df.iterrows():
      x,y,z=row['a'],row['b'],row['c']
      temp_values=[quadratic_weather_model(t,x,y,z) for t in time_values]
      plt.plot(time_values,temp_values,marker='o',linestyle='-',label=f'a=[x],b=[y],c=[z]')
    plt.title('temperature variation')
    plt.xlabel('time')
    plt.ylabel('Temperature[(°C)]')
    plt.grid(True)
    plt.xlim(0,10)
    plt.legend()
    plt.show()
  except FileNotFoundError:
    print("File not found")
main()
```
